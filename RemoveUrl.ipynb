```python
import string,time
```

```python
exclude=string.punctuation
```

# Manual excluding

```python
punctuation_sample = """
Hello!!! How are you? I'm doing great, thanks for asking!
Well, actually... I'm not sure; I've been feeling kind of weird lately.
"Life is what happens when you're busy making other plans," said John Lennon.
Wait—did you hear that? It sounded like a crash—or maybe it was just thunder?
The price is $19.99; that's 20% off—what a deal!!!
She said: "I'll be there at 5:30 p.m. sharp—don't be late!"
The items include: apples, bananas, oranges; and also: grapes, watermelon, and strawberries.
"Can you believe it?" she asked. "He actually said, 'I don't care!'"
I like coffee, tea, and juice; but my favorite is espresso—no question about it.
The meeting is at 3:00 PM (please be on time) — and don't forget your ID badge.
What? You want to go? But we just got here... fine, let's go!
I bought 3 apples, 2 oranges, and 5 bananas—plus 1 watermelon (total: 11 fruits).
He said: "I'm not sure if I can come; I have a meeting at 4:30 p.m."
"Really?" she replied. "That's too bad... I was hoping you'd join us!"
Note: Please bring the following items: (1) your laptop, (2) charger, (3) notebook; and (4) pen.
The temperature reached 95°F—that's 35°C—which is really hot!!!
She asked, "Have you seen my keys?" and I said, "No, I haven't—sorry!"
The recipe calls for: flour, sugar, eggs (2 large), butter (1/2 cup), and milk.
Wow!!! This is amazing—I can't believe it—truly incredible!!!
P.S. Don't forget to bring snacks—and maybe some drinks too?
"""
```

```python
def remove_punc(text):
  for char in exclude:
    text=text.replace(char,"")
  return text
```

```python
start=time.time()
print(remove_punc(punctuation_sample))
time1=time.time()-start
print(time1)
```

# standard way

```python
def remove_punc1(text):
  return text.translate(str.maketrans('','',exclude))
```

```python
start=time.time()
print(remove_punc1(punctuation_sample))
time1=time.time()-start
print(time1)
```
