```python
url_sample = """
Visit our website at https://www.example.com for more information.
You can also check out http://example.org or https://sub.domain.co.uk/page?q=test.
For documentation, go to https://docs.python.org/3/library/re.html?highlight=regex#re.sub.
Our social media: https://twitter.com/username, https://www.facebook.com/profile.php?id=12345, and https://www.instagram.com/p/CxYz123AbC/.
For images: http://images.example.com/photo.jpg and https://cdn.site.com/uploads/2024/image.png?width=200&height=150.
You can also use short URLs like https://bit.ly/3xYzABC or https://tinyurl.com/2x9k4m.
Email us at info@example.com or support@company.co.uk.
FTP server: ftp://files.example.com/documents/report.pdf.
For local development: http://localhost:8080/api/v1/users or https://127.0.0.1:3000/dashboard.
Subdomains: https://blog.mywebsite.com, https://shop.example-store.com, http://test.dev.app.example.com.
Complex URLs: https://example.com/search?q=python+programming&lang=en&page=2#results.
Using different protocols: http://example.com, https://example.com, ftp://example.com, file:///C:/Users/Documents/file.txt.
Social handles: @username, #hashtag, and check www.example-page.com without protocol.
"""
```

```python
import re
def remove_url(text):
  p=re.compile(r'https?://\S+|www\.S+')
  return p.sub('',text)
```

```python
remove_url(url_sample)
```
