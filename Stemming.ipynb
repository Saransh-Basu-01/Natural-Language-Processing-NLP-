```python
import nltk
```

```python
from nltk.stem.porter import PorterStemmer
```

```python
ps=PorterStemmer()
def stem_words(text):
  return " ".join([ps.stem(word) for word in text.split()])
```

```python
def stem_words1(text):
  new_text=[]
  for word in text.split():
    new_text.append(ps.stem(word))
  return " ".join(new_text)
```

```python
sample="walk walks walking walked"
stem_words(sample)
```

```python
sample="walk walks walking walked"
stem_words1(sample)
```

```python
short_stemming_sample = """
The runners were running quickly. They had run many miles.
The leaves were falling gently. The fallen leaves covered the ground.
The children played games. Their playing continued until dark.
The teacher taught history. She had taught the same lesson.
The students learned quickly. Their learning never stopped.
The fisherman caught fish. He had fished for years.
The birds were flying high. They flew over the mountains.
The flowers bloomed in spring. They had bloomed every year.
The cars drove fast. Their driving was excellent.
The workers worked hard. They had worked for decades.
The water flowed down the river. It had flowed for centuries.
The sun shone brightly. It had shined on this land forever.
"""
```

```python
stem_words(short_stemming_sample)
```
