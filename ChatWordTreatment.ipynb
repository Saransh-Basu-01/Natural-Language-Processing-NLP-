```python
import re
def remove_slang(text):
  new_text=[]
  for word in text.split():
    clean_word = re.sub(r'[^\w]', '', word)
    if clean_word.upper() in slang_dict:
      new_text.append(slang_dict[clean_word.upper()])
    else:
      new_text.append(word)
  return " ".join(new_text)
```
