```python
sample_text = '<html><head><title>Sample HTML Document</title><meta charset="UTF-8"><style>body { font-family: Arial, sans-serif; }.highlight { background-color: yellow; }</style></head><body><h1>Welcome to My Sample Page</h1><h2>Hello my name is John Doe</h2><h3>And this is a sample document</h3><p>This is a <strong>paragraph</strong> with <em>various</em> <u>formatting</u> tags.</p><div class="container"><p>Here\'s a list of items:</p><ul><li>First item with <a href="https://example.com">a link</a></li><li>Second item with <span class="highlight">highlighted text</span></li><li>Third item with <b>bold</b> and <i>italic</i> text</li></ul></div><table border="1"><tr><th>Name</th><th>Age</th><th>City</th></tr><tr><td>Alice</td><td>25</td><td>New York</td></tr><tr><td>Bob</td><td>30</td><td>London</td></tr></table><p>Here\'s a <a href="#" onclick="alert(\'Hello!\')">link with JavaScript</a>.</p><img src="image.jpg" alt="Sample image" width="200" height="150"><blockquote>This is a blockquote with some <cite>citation</cite> text.</blockquote><p>Some inline tags: <code>code</code>, <sub>subscript</sub>, <sup>superscript</sup>, <mark>marked text</mark>, and <small>small text</small>.</p><form action="/submit" method="POST"><label for="name">Name:</label><input type="text" id="name" name="name" value="John"><br><label for="email">Email:</label><input type="email" id="email" name="email" value="john@example.com"><br><input type="submit" value="Submit"></form><footer><p>&copy; 2024 My Company. All rights reserved.</p></footer></body></html>'
```

```python
sample_text
```

```python
import re
def striphtml(text):
  p=re.compile(r'<.*?>')
  return p.sub('',text)
```

```python
striphtml(sample_text)
```
